```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv('/content/housepricedata.csv')
df
```

```python
df.head()
```

```python
#to check the feature names
x=df.iloc[:,0:10]
print(x)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
#scale the data
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler(feature_range=(0,1))
x_scale=scaler.fit_transform(x)
print(x_scale)
```

```python
y=df.iloc[:,10]
print(y)
```

```python
#split the dataset
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x_scale,y,test_size=0.2)
```

```python
#Build the ANN model
model=Sequential()
```

```python
#input layer
input_layer=Dense(20,input_shape=(10,))
```

```python
#hidden layer1
hidden_layer1=Dense(20,activation='relu')
model.add(hidden_layer1)
```

```python
#hidden layer2
hidden_layer2=Dense(20,activation='relu')
model.add(hidden_layer2)
```

```python
#output layer
output_layer=Dense(1,activation='sigmoid')
model.add(output_layer)
```

```python
model.compile(loss='binary_crossentropy',optimizer='sgd',metrics=['accuracy'])
```

```python
model.fit(x_train,y_train,epochs=50,batch_size=10)
```

```python
val_loss,val_acc=model.evaluate(x_test,y_test)
print("Loss:",val_loss)
print("Accuracy:",val_acc)
```

```python
#prediction
l=[8450,7,5,856,2,1,3,8,0,548]
```

```python
scaled_l=scaler.transform([l])
print(scaled_l)
```

```python
result=model.predict(scaled_l)
print(result)
if result>0.5:
  print("House price is above average level")
else:
  print("House price is below average level")
```
