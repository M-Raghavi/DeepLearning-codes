```python
#Back Propagation Neural Network
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
```

```python
iris = load_iris()
x = iris.data
y = iris.target
```

```python
y = pd.get_dummies(iris.target).values
y = pd.get_dummies(iris.target, dtype='int').values
print(y)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=20, random_state=4)
```

```python
print(x_train.size)
print(x_test.size)
print(y_train.size)
print(y_test.size)
```

```python
learning_rate = 0.1
iterations = 1000
N = y_train.size
input_size = 4
hidden_size = 2
output_size = 3
results = pd.DataFrame(columns=["mse", "accuracy"])
print(results)
```

```python
## Numpy random number generator for getting the same sequence of numbers.
##The number 10 serves as the seed value for the random nunber generator in numpy
np.random.seed(10)
## Initializing the weight for the hidden layer
## The below code generates a weight matrix w1
w1 = np.random.normal(scale=0.5, size=(input_size, hidden_size))
'''This part generates random numbers from a normal distribution with a mean of 0 and a standard deviation of 0.5'''
'''Overall, this line initializes the weights of the connections between the input layer and the hidden layer of a neural network with random values drawn from a normal distribution with a standard deviation of 0.5'''
print(w1)
print()
## Now, initializing the weight for the output layer
## the below code generates a weight matrix w2e
w2 = np.random.normal(scale=0.5, size=(hidden_size, output_size))
print(w2)
```

```python
## Helper functions: sigmoid, mse and accuracy
def sigmoid(x):
  return 1 / (1 + np.exp(-x))
```

```python
## 'y_pred' contains the predicted probabilities for each class for each sample
## 'y_true' contains the true class label for each sample, typically in one-hot encoded format
def mean_squared_error(y_pred, y_true):
  squarred_erorr = (y_pred - y_true) ** 2
  sum_squared_error = squarred_erorr.sum()
  mse = sum_squared_error / (2 * y_pred.size)
  return mse
```

```python
## The below code calculates the accuracy between predicted y_pred values and true y_true values
def accuracy(y_pred, y_true):
  ## Compare  the index of the maximum value in predicted and true arrays
  acc = y_pred.argmax(axis=1) == y_true.argmax(axis=1)
  ## Compute the mean of the boolean array obtained
  return acc.mean
## y_pred.argmax(axis=1): this returns the index of the maximum value in the predicted array 'y_pred'
## y_true.argmax(axis=1): this returns the index of the maximum value in the true array 'y_true'
```

```python
## Feedforward
Z1 = np.dot(x_test, w1)
A1 = sigmoid(Z1)
Z2 = np.dot(A1, w2)
A2 = sigmoid(Z2)
acc = accuracy(A2, y_test)
print("Accuracy: ", acc)
```

```python
#Backpropagation Neural Network
for i in range(iterations): #iterations=5000
  #Feedforward propagation on hidden layer
  Z1 = np.dot(x_train, w1)
  A1 = sigmoid(Z1)
  #Feedforward
  Z2 = np.dot(A1, w2)
  A2 = sigmoid(Z2)
  #Calculating error
  mse=mean_squared_error(A2,y_train)
  acc=accuracy(A2,y_train)
  #results=resultd.append({"mse":mse,"accuracy":acc},ignore_index=True)
  results.loc[i]={"mse":mse,"accuracy":acc}
  print(results)
  #backpropagation
  E1=A2-y_train
  dw1=E1*A2*(1-A2)
  print("dw1")
  print(dw1)
  print(dw1.size)
  print()
  E2=np.dot(dw1,w2.T)
  dw2=E2*A1*(1-A1)
  print("dw2")
  print(dw2)
  print(dw2.size)
  print()
  #weights updates
  w2_update=np.dot(A1.T,dw1)/N
  w1_update=np.dot(x_train.T,dw2)/N
  w2=w2-learning_rate*w2_update
  w1=w1-learning_rate*w1_update
  print()
  print('w2_update',w2_update) #learning_rate = 0.1
  print('w1_update',w1_update)
  print('w2',w2)
  print('w1',w1)
```
